```python
import pandas as pd
import statistics as stats
```

```python
low_range, high_range = [], []

for i in range(15):
    if i < 7:
        low_range.append([i, i * 2])
        high_range.append([i, i * 5])
    else:
        low_range.append([i, 10 - i])
        high_range.append([i, 60 - i * 5])
```

```python
df_low = pd.DataFrame(low_range, columns=['x', 'y'])
```

```python
df_low.plot.scatter(x='x', y='y')
```

```python
df_high = pd.DataFrame(high_range, columns=['x', 'y'])
```

```python
df_high.plot.scatter(x='x', y='y')
```

```python
stats.stdev(df_low.x)
```

```python
stats.stdev(df_high.y)
```
